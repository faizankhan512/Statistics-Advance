14.How does the Central Limit Theorem relate to the normal distribution?

ANS - Relationship Between the Central Limit Theorem (CLT) and the Normal Distribution

The CLT says that the distribution of the sample mean (or sum) of a large number of independent, identically distributed random variables approaches a normal distribution, regardless of the original data‚Äôs distribution.

In other words, no matter what the shape of your original data (skewed, uniform, etc.), the distribution of the average of many samples will look approximately normal if the sample size is large enough.

Why does this matter?
It lets us use the powerful properties and tools of the normal distribution (like Z-scores, confidence intervals) in many situations, even when the original data isn‚Äôt normal.

So, the CLT justifies the widespread use of the normal distribution in statistics.

18.What is the significance of confidence intervals in statistical analysis?

ANS - Significance of Confidence Intervals in Statistical Analysis

1. Measure of Precision and Uncertainty
Confidence intervals (CIs) provide a range of plausible values for an unknown population parameter (e.g., mean), showing the precision of the estimate and accounting for sampling variability.

2. More Informative than Point Estimates
Unlike a single point estimate, CIs show how reliable or uncertain that estimate is, helping to understand the possible spread of the true value.

3. Basis for Inference
CIs allow researchers to make probabilistic statements, e.g.,
‚ÄúWe are 95% confident the true mean lies between X and Y.‚Äù
This guides decision-making and hypothesis testing.

4. Helps in Comparing Groups
If confidence intervals for two groups do not overlap, it suggests a significant difference between their parameters.

5. Supports Reproducibility
CIs quantify the expected variability in repeated samples, helping to interpret results consistently across studies.

22.What is the concept of expected value in a probability distribution?

ANS -  Definition:
The expected value (or mean) of a random variable is the long-run average value it takes after many repetitions of the experiment. It represents the center or balance point of the distribution.

Interpretation:

* Expected value tells you the average outcome you can expect if you repeat the process many times.

* It may not always be a value the random variable can actually take (e.g., expected value of a dice roll is 3.5).

```python
# 3.Write a Python function to calculate the probability distribution function
# (PDF) of a Bernoulli distribution?

def bernoulli_pmf(x, p):
    """
    Calculate the Bernoulli PMF.

    Parameters:
    - x: outcome (0 or 1)
    - p: probability of success (0 <= p <= 1)

    Returns:
    - Probability of observing x
    """
    if x not in [0, 1]:
        return 0  # Bernoulli only takes 0 or 1
    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.7  # Probability of success
print("P(X=1):", bernoulli_pmf(1, p))
print("P(X=0):", bernoulli_pmf(0, p))
```

```python
#6.Write a Python program to calculate and plot the cumulative distribution
# function (CDF) of a discrete uniform distribution?

import matplotlib.pyplot as plt

# Define the discrete uniform distribution values (e.g., 1 to 6)
values = list(range(1, 7))

# Each value has equal probability
probabilities = [1/len(values)] * len(values)

# Calculate the CDF values (cumulative sum of probabilities)
cdf = []
cumulative = 0
for p in probabilities:
    cumulative += p
    cdf.append(cumulative)

# Plot the CDF
plt.step(values, cdf, where='post')
plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.ylim(0, 1.05)
plt.grid(True)
plt.show()
```

```python
# 10.Implement the Central Limit Theorem (CLT) using Python for a non-normal
# distribution?

import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_param = 1  # rate parameter for exponential distribution
sample_size = 30  # size of each sample
num_samples = 1000  # number of samples

# Step 1: Generate sample means
sample_means = []
for _ in range(num_samples):
    sample = np.random.exponential(1/lambda_param, sample_size)
    sample_means.append(np.mean(sample))

# Step 2: Plot histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='skyblue')
plt.title(f'Distribution of Sample Means (n={sample_size})')
plt.xlabel('Sample Mean')
plt.ylabel('Density')

# Optional: overlay normal distribution curve for comparison
from scipy.stats import norm
mean_of_means = np.mean(sample_means)
std_of_means = np.std(sample_means)
x = np.linspace(min(sample_means), max(sample_means), 100)
plt.plot(x, norm.pdf(x, mean_of_means, std_of_means), 'r--', label='Normal PDF')
plt.legend()

plt.show()
```

```python
#19.Implement hypothesis testing using Z-statistics for a sample dataset?

import numpy as np
from scipy.stats import norm

# Sample data
sample = [85, 78, 90, 88, 76, 95, 89, 84, 91, 87]

# Known population parameters
population_mean = 80
population_std = 10  # known population standard deviation

# Sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)

# Calculate Z-statistic
z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z)))

# Output results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z:.3f}")
print(f"P-value = {p_value:.4f}")

# Interpretation at alpha = 0.05
if p_value < 0.05:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

```python
# 20. Create a confidence interval for a dataset using Python and interpret the
# result?

import numpy as np
from scipy import stats

# Sample dataset
data = [85, 78, 90, 88, 76, 95, 89, 84, 91, 87]

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
n = len(data)
confidence_level = 0.95

# Calculate the t-critical value (two-tailed)
t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n-1)

# Calculate margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Calculate confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"{int(confidence_level*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Interpretation
print(f"\nInterpretation: We are {int(confidence_level*100)}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")
```

```python
# 27.Implement Z-score normalization on a dataset and visualize its
# transformation?

import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
data = [85, 78, 90, 88, 76, 95, 89, 84, 91, 87]

# Calculate mean and standard deviation
mean = np.mean(data)
std = np.std(data)

# Apply Z-score normalization
z_scores = [(x - mean) / std for x in data]

# Plot original data vs normalized data
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=5, color='salmon', edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# 28.Write a Python function to calculate the skewness and kurtosis of a dataset
# generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def skewness_kurtosis_normal_data(mean=0, std=1, size=1000):
    # Generate data from a normal distribution
    data = np.random.normal(mean, std, size)

    # Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # by default, Fisher‚Äôs definition (excess kurtosis)

    return data_skewness, data_kurtosis

# Example usage
skew_val, kurt_val = skewness_kurtosis_normal_data()
print(f"Skewness: {skew_val:.4f}")
print(f"Kurtosis (excess): {kurt_val:.4f}")
```
